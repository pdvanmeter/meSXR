```python
import gc
import os
import cPickle as pickle
import numpy as np
import matplotlib.pyplot as plt
import mesxr.calibration.trimscan as calib
import mesxr.calibration.plots as plot
import mesxr.operation.configuration as cfg

%matplotlib inline
```

```python
save_path = '/home/pdvanmeter/data/meSXR/MST/calibration/pickles'
calibration_name = 'MST_midE.pkl'

# Load in the detector object - expect this to take a couple of minutes
file = open(os.path.join(save_path, calibration_name), 'rb')
gc.disable()
p3_det = pickle.load(file)
gc.enable()
file.close()
```

## 1 - ppcd_8_color

```python
config_dir = '/home/pdvanmeter/analysis/python/mesxr_dev/configurations/midE'
config_name = 'ppcd_8_color'

# Load the threshold map
thresholds = [4.0, 4.2, 4.4, 4.6, 4.8, 5.0, 5.5, 6.0]
threshold_map = cfg.ME_horiz_stripes_map(p3_det, thresholds)

# Configure the detector
cfg.configure_detector(p3_det, threshold_map, config_dir, config_name=config_name)
```

```python
fig = plt.figure(1, figsize=(8,6), dpi=110)
ax = fig.add_subplot(1, 1, 1)
plt.tick_params(labelsize=14)

cax = ax.imshow(threshold_map.T, vmin=4.0, vmax=6.0)
fig.colorbar(cax, orientation='horizontal', label='Threshold (keV)')
ax.set_title(r'Threshold map ({0:})'.format(config_name))
```
